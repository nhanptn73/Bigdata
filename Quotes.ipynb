```python
import requests
from bs4 import BeautifulSoup

base_url = 'http://quotes.toscrape.com/page/'
main_url = 'http://quotes.toscrape.com'
```

```python
#cau1.2b
def author_name(result):
  for t in result:
    #print(t.span.text)
    a = t.find("small", class_="author")
    author = a.text
  return author
```

```python
import csv
def tacgiaLink(html):  
  output = []

  soup = BeautifulSoup(html)
  tag = soup.find_all("div", class_="quote")
  print(tag)
  for t in tag:
    #print quote
    text_out =  t.span.text
    quotes=text_out
    #print(t.span.text)
    #print author
    a = t.find("small", class_="author")
    author = a.text
    #text_out = text_out + ',"' + author + '"' 
    #print("By : ", author)
    #print link author
    hrefs = t.a
    link = hrefs.get('href')
    link_url = main_url+ link
    #print(link_url)
    #print date born author
    response_auth = requests.get(link_url)
    html_auth = response_auth.content
    auth_soup = BeautifulSoup(html_auth)
    born_soup = auth_soup.find("span", class_="author-born-date")
    born = born_soup.text
    #print("Author DOB:", born.text)
    #text_out = text_out + ',"' + born + '"' 
    #output.append(text_out)
    #write csv
    print(author,link_url,born,quotes)
    with open('Quotes.csv', 'a') as f:
      writer = csv.writer(f)
      writer.writerow([author,link_url,born,quotes])
```

```python
import requests
from bs4 import BeautifulSoup
for i in range(1,11):
  url = base_url + str(i)
  response = requests.get(url)
  html = response.content
#cau1.2a
  soup = BeautifulSoup(html)
  result = soup.find_all("div", class_="quote")
  print(result)
#cau1.1 save file txt
  with open('kq.txt','ab') as fo:
    fo.write(html)
  fo.close();
  tacgiaLink(html);
```
